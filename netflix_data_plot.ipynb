```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading csv file and cleaning dataset

df = pd.read_csv('netflix_data.csv')

df = df.dropna(subset=['type', 'release_year', 'rating', 'country', 'duration'])

```

```python
# plotting bar graph of movies vs TV shows

type_counts = df['type'].value_counts()  # will count the number of categories in column 'type'.

plt.bar(type_counts.index, type_counts.values)
plt.ylabel('Number of movies and shows')
plt.plot()
plt.savefig('movies_vs_TVshows')
plt.show()
```

```python
rating_counts = df['rating'].value_counts()
print(rating_counts)

plt.pie(rating_counts,labels = rating_counts.index, autopct='%1.1f%%')

plt.xticks(rotation = 90)
plt.plot()
plt.savefig('Division of content accordin to rating')
plt.show()
```

```python
movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_int'] = movie_df['duration'].str.replace('min', '').astype(int)

plt.hist(movie_df['duration_int'], bins=30, color = 'purple', edgecolor = 'black')
plt.title('Distribtution fo Movie Duration')
plt.xlabel('Duration in minutes')
plt.ylabel('Number of movies')

plt.savefig('Distribution according to Duration in minutes')
plt.show()
```

```python
release_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(16,9))
plt.tight_layout()
plt.grid()
plt.xticks(ticks=range(release_counts.index.min(), release_counts.index.max()+1, 3))
plt.scatter(release_counts.index, release_counts.values)
plt.xlabel('Year in which released')
plt.ylabel('Number of movies and shows')

plt.savefig('Number of movies released in each year')
plt.show()
```
